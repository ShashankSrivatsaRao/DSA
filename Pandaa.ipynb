```python
df1 = pd.DataFrame({
    'id':[1,2,3,4],
    'name':['Sandeep','Meghana','Nayana','Darshan']
})

df2=pd.DataFrame({
    'id':[2,3,4,5],
    'course':['Gen','CPP','HLD','LLD']
})
#Doin Inner Join as if in SQL
inner_result =pd.merge(df1,df2,how='inner',on='id')
inner_result
```

```python
#Doing Outer Join as if in the SQL Outer Join --null is replaced by NaN
outer_result =pd.merge(df1,df2,how='outer',on='id')
outer_result
```

```python
left_result =pd.merge(df1,df2,how='left',on='id')
left_result
```

```python
#Right JOIN 
right_result =pd.merge(df1,df2,how='right',on='id')
right_result
```

```python
##suppose we do not have the same name for the column in both the data frame
ff1= pd.DataFrame({
    'id1':[1,2,3,4],
    'name':['Dar','Suj','Sur','Sha']
})
ff2= pd.DataFrame({
    'id2':[2,3,4,5],
    'team':['EC','IT','DEV','DevOps']
})

inner_join = pd.merge(ff1,ff2,how='inner',left_on='id1',right_on='id2')
inner_join
```

```python
#Concatination using concat like union in sql
import pandas as pd
ff1= pd.DataFrame({
    'id1':[1,2,3,4],
    'name':['Dar','Suj','Sur','Sha']
})
ff2= pd.DataFrame({
    'id2':[2,3,4,5],
    'team':['EC','IT','DEV','DevOps']
})

join = pd.concat([ff1,ff2],axis=0) #axis=0 means stacked vertically together
oin=pd.concat([ff1,ff2],axis=0,ignore_index=True) #ignoring the index
join
oin
```

```python
#vjoin=pd.concat([ff1,ff2],axis=1) #Horizontal Concatination
vjoin=pd.concat([ff1,ff2],axis=0,keys=['ff1','ff2']) #Getting the hierarchy
vjoin
```
